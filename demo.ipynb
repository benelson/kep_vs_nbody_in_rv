```python
import numpy as np
import sys

### To install, 'pip install rebound'
import rebound
### To install, 'pip install radvel'
import radvel

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)

### The modules below allow us to create a progress bar for our for loops
### To install, 'pip install ipywidgets'
from ipywidgets import FloatProgress
from IPython.display import display
```

Let's define some constants and an array of times to evaluate our RV models at.

```python
AUperTwoPiYearToMetersPerSecond = 4743.72 * 2 * np.pi
DaysInYear = 365.2425
DaysToTwoPiYears = 2.*np.pi/DaysInYear
MsunToMJupiter = 1047.56

times = np.linspace(0., DaysInYear*2, 2000) # 2000 observations spaced equally across a 2 year baseline
```

### Define your planetary system here.

The example below is a 2-planet system around a 1 solar mass star. We'll start off with very low-amplitude (thus, very low-mass) "planets" to test that the Keplerian and n-body models match.

```python
### mass of the central star in [solar masses]
mstar = 1.

### orbital periods in [days]
P = [20., 42.]

### semi-amplitudes in [m/s]
K = [0.001, 0.001]

### eccentricities
e = [0.08, 0.04] 

### argument of pericenter in [radians]
omega = [np.e, np.pi] 

### time of periastron passage in [days]
T = [1., 2.] 
```

Now, we'll define some functions to generate RV data based on the above system properties and a list of observation times.

Doing this with $\tt{radvel}$ is fairly simple. $\tt{radvel.kepler.rv\_drive}$ will evaluate the RV signal at each observation time.

Unfortunately, we'll have to improvise with $\tt{Rebound}$ since it requires physical masses to integrate orbits. The $\tt{radvel.orbit.msini}$ method will get us close, but only if the mass of the planet is much less than star. Also note that Rebound works in units of 1 year = $2\pi$.

```python
def rvs_from_keplerian_model(mstar=1., P=[365.], K=[2.], e=[0.04], omega=[0.], T=[0.], times=times):
    assert len(P)==len(K)==len(e)==len(omega)==len(T)
    npl = len(P)
    
    rvs = np.zeros(len(times))
    
    for p in range(npl):
        rvs += radvel.kepler.rv_drive(times, [P[p], T[p], e[p], omega[p], K[p]])
    return rvs

def rvs_from_nbody_model(mstar=1., P=[365.], K=[2.], e=[0.04], omega=[0.], T=[0.], times=times):
    assert len(P)==len(K)==len(e)==len(omega)==len(T)
    npl = len(P)
    
    times_2pi = times * DaysToTwoPiYears
    T_2pi = np.array(T) * DaysToTwoPiYears
    
    sim = rebound.Simulation()
    sim.add(m=mstar)
    for p in range(npl):
        m = radvel.orbit.Msini(K[p], P[p], mstar, e[p], Msini_units='jupiter')/MsunToMJupiter
        sim.add(P=P[p]*DaysToTwoPiYears, m=m, e=e[p], omega=omega[p], T=T_2pi[p])
    sim.move_to_com()

    rvs = np.zeros(len(times))

    for i, t in enumerate(times_2pi):
        sim.integrate(t)
        rvs[i] = -sim.particles[0].vy * AUperTwoPiYearToMetersPerSecond
        
    return rvs
```

Now we can compute RVs based on a Keplerian model and n-body model.

```python
rvs_kep = rvs_from_keplerian_model(mstar=mstar, P=P, K=K, e=e, omega=omega, T=T, times=times)
rvs_nbody = rvs_from_nbody_model(mstar=mstar, P=P, K=K, e=e, omega=omega, T=T, times=times)
```

The moment of truth. Hopefully the RVs match. We can either overplot both RV curves or look at the difference between them (i.e., $\tt{rvs\_kep - rvs\_nbody}$).

```python
plt.plot(times, rvs_kep, 'x')
plt.plot(times, rvs_nbody, '+')
#plt.plot(times, rvs_kep - rvs_nbody, 'o')
plt.xlabel('Time [days]', fontsize=16)
plt.ylabel('RV [m/s]', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

These models differ by a couple $\mu$m/s, roughly 1000 times less than the RV amplitudes. Good enough for our purposes!

Now let's increase the masses to something more realistic and rerun the same models.

```python
K = [5., 5.]

rvs_kep = rvs_from_keplerian_model(mstar=mstar, P=P, K=K, e=e, omega=omega, T=T, times=times)
rvs_nbody = rvs_from_nbody_model(mstar=mstar, P=P, K=K, e=e, omega=omega, T=T, times=times)

plt.plot(times, rvs_kep, 'x')
plt.plot(times, rvs_nbody, '+')
#plt.plot(times, rvs_kep - rvs_nbody, 'o')
plt.xlabel('Time [days]', fontsize=16)
plt.ylabel('RV [m/s]', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

For this particular set of parameters, the worst the models differ is ~0.25 m/s, which is well below today's typical RV precision. This suggests that a Keplerian model would be a good approximation for this system.

But this difference definitely appears to grow with time. Let's extend the observational baseline to 10 years and see how bad it gets.

```python
baseline = DaysInYear*10
times_ext = np.linspace(0., baseline, 2000)

rvs_kep_ext = rvs_from_keplerian_model(mstar=mstar, P=P, K=K, e=e, omega=omega, T=T, times=times_ext)
rvs_nbody_ext = rvs_from_nbody_model(mstar=mstar, P=P, K=K, e=e, omega=omega, T=T, times=times_ext)
```

```python
plt.plot(times_ext, rvs_kep_ext, 'x')
plt.plot(times_ext, rvs_nbody_ext, '+')
#plt.plot(times_ext, rvs_kep_ext - rvs_nbody_ext, 'o')
plt.xlabel('Time [days]', fontsize=16)
plt.ylabel('RV [m/s]', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

After 10 years, the difference has grown to over 1 m/s, which could be potentially detectable.

Let's say we're able to discriminate between a Keplerian and n-body model at the 2 m/s level. We could find the time at which this threshold is achieved. And what if we want to understand how this critical time changes across a $\textit{range}$ of orbital periods?

We're going to scale up our operation ~400-fold! Let's evaluate our Keplerian and n-body models across a range of orbital periods and find the time at which we can detect differences in these models.

```python
rv_crit = 2
step = 0.5

P1_list = np.arange(15,25,step)
P2_list = np.arange(37,47,step)

grid_vals = []

### To keep track of progress, ipywidgets has a nice loading bar.
### If you don't import this module, just comment these two lines out.
f = FloatProgress(min=0, max=len(P1_list))
display(f)

### Loop over orbital periods in P1_list and P2_list
for i,P1 in enumerate(P1_list):
    for P2 in P2_list:
        rvs_kep = rvs_from_keplerian_model(mstar=mstar, P=[P1,P2], K=K, e=e, omega=omega, T=T, times=times_ext)
        rvs_nbody = rvs_from_nbody_model(mstar=mstar, P=[P1,P2], K=K, e=e, omega=omega, T=T, times=times_ext)
        diff = rvs_kep - rvs_nbody

### Find the indices where Keplerian/nbody difference greater than rv_crit
        inds = np.where(diff > rv_crit)[0]

### Find the corresponding first time that the Keplerian/body difference is achieved
        if inds.size > 0:
            t_crit = times_ext[inds[0]]
        else:
            t_crit = baseline
        
        grid_vals.append([P1, P2, t_crit])

### Comment this out if you don't have FloatProgress imported
    f.value = i+1

#        print str(100.*(i+1)/len(P1_list)) + "% done..."
```

Now we'll plot up the results on a colored grid.

```python
x_list, y_list, z_list = np.transpose(grid_vals)

N = int(len(z_list)**.5)
z = z_list.reshape(N, N).T[::-1]

plt.imshow(z, cmap='viridis_r', \
           extent=(min(x_list)-step/2., max(x_list)+step/2., min(y_list)-step/2., max(y_list)+step/2.))

plt.xlabel("P1 [days]", fontsize=16)
plt.ylabel("P2 [days]", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.colorbar().set_label("days until detectable 'kep - nbody' difference")
plt.show()
```

The horizontal and vertical axis are the values in $\tt{P1\_list}$ and $\tt{P2\_list}$ respectively. The color corresponds to the time at which a model difference of $\tt{rv\_crit}$ is achieved.

The main feature is clearly the slope tracing a 2:1 period commensurability. But it's not perfectly symmetric. As orbital periods separate, the time to notice a difference gets larger until it's no longer noticeable (dark blue region on left). However for period ratios interior to 2:1 (i.e., planets getting closer together), there are detectable n-body effects even if the planets aren't orbiting in a mean-motion resonance.

There appears to be another dynamically active feature in the lower right hand corner, where n-body effects are typically detected much later. This corresponds to a 5:3 period commensurability. There's a gap between the 2:1 and 5:3 bands where n-body effects aren't detectable.
